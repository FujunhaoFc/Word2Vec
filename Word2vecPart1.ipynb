```python
labeled_train = pd.read_csv("word2vec/labeledTrainData.tsv", header=0, delimiter="\t", quoting=3)
unlabeled_train = pd.read_csv("word2vec/unlabeledTrainData.tsv", header=0, delimiter="\t", quoting=3)
test = pd.read_csv("word2vec/testData.tsv", header=0, delimiter="\t", quoting=3)
```

```python
from bs4 import BeautifulSoup

# Initialize the BeautifulSoup object on a single movie review
example1 = BeautifulSoup(labeled_train["review"][0])

# Print the raw review and then the output of get_text(), for
# comparison
print(labeled_train["review"][0])
print(example1.get_text())
```

```python
import re
# Use regular expressions to do a find-and-replace
letters_only = re.sub("[^a-zA-Z]",           # The pattern to search for
                      " ",                   # The pattern to replace it with
                      example1.get_text() )  # The text to search
print(letters_only)
```

```python
lower_case = letters_only.lower()        # Convert to lower case
words = lower_case.split()               # Split into words
```

```python
import nltk
nltk.download("stopwords")
```

```python
from nltk.corpus import stopwords # Import the stop word list
print(stopwords.words("english"))
```

```python
# Remove stop words from "words"
words = [w for w in words if not w in stopwords.words("english")]
print(words)
```

```python
def review_to_words( raw_review ):
    # Function to convert a raw review to a string of words
    # The input is a single string (a raw movie review), and
    # the output is a single string (a preprocessed movie review)
    #
    # 1. Remove HTML
    review_text = BeautifulSoup(raw_review).get_text()
    #
    # 2. Remove non-letters
    letters_only = re.sub("[^a-zA-Z]", " ", review_text)
    #
    # 3. Convert to lower case, split into individual words
    words = letters_only.lower().split()
    #
    # 4. In Python, searching a set is much faster than searching
    #   a list, so convert the stop words to a set
    stops = set(stopwords.words("english"))
    #
    # 5. Remove stop words
    meaningful_words = [w for w in words if not w in stops]
    #
    # 6. Join the words back into one string separated by space,
    # and return the result.
    return( " ".join( meaningful_words ))
```

```python
clean_labeled_review = review_to_words( labeled_train["review"][0] )
clean_unlabeled_review = review_to_words( unlabeled_train["review"][0] )
print(clean_labeled_review)
print(clean_unlabeled_review)
```

```python
# Get the number of reviews based on the dataframe column size
num_labeled_reviews = labeled_train["review"].size

# Initialize an empty list to hold the clean reviews
clean_labeled_train_reviews = []

# Loop over each review; create an index i that goes from 0 to the length
# of the movie review list
for i in range( 0, num_labeled_reviews ):
    # If the index is evenly divisible by 1000, print a message
    if( (i+1)%1000 == 0 ):
        print("Review %d of %d\n" % ( i+1, num_labeled_reviews ))
    clean_labeled_train_reviews.append( review_to_words( labeled_train["review"][i] ))
```

```python
clean_labeled_train_reviews[0:5]
```

```python
num_unlabeled_reviews = unlabeled_train["review"].size

clean_unlabeled_train_reviews = []

for i in range( 0, num_unlabeled_reviews ):
    if( (i+1)%1000 == 0 ):
        print("Review %d of %d\n" % ( i+1, num_unlabeled_reviews ))
    clean_unlabeled_train_reviews.append( review_to_words( unlabeled_train["review"][i] ))
```

```python
clean_unlabeled_train_reviews[0:5]
```

```python
print("Creating the bag of words...\n")
from sklearn.feature_extraction.text import CountVectorizer

# Initialize the "CountVectorizer" object, which is scikit-learn's bag of words tool.
vectorizer = CountVectorizer(analyzer = "word",   \
                             tokenizer = None,    \
                             preprocessor = None, \
                             stop_words = None,   \
                             max_features = 5000)

# fit_transform() does two functions: First, it fits the model
# and learns the vocabulary; second, it transforms our training data
# into feature vectors. The input to fit_transform should be a list of
# strings.
labeled_train_data_features = vectorizer.fit_transform(clean_labeled_train_reviews)

# Numpy arrays are easy to work with, so convert the result to an array
labeled_train_data_features = labeled_train_data_features.toarray()
```

```python
labeled_train_data_features
```

```python
print(labeled_train_data_features.shape)
```

```python
# Take a look at the words in the vocabulary
vocab = vectorizer.get_feature_names_out()
print(vocab)
```

```python
# Sum up the counts of each vocabulary word
dist = np.sum(labeled_train_data_features, axis=0)

# For each, print the vocabulary word and the number of times it
# appears in the training set
for tag, count in zip(vocab, dist):
    print(count, tag)
```

```python
print("Training the random forest...")
from sklearn.ensemble import RandomForestClassifier

# Initialize a Random Forest classifier with 100 trees
forest = RandomForestClassifier(n_estimators = 100)

# Fit the forest to the training set, using the bag of words as
# features and the sentiment labels as the response variable
#
# This may take a few minutes to run
forest = forest.fit( labeled_train_data_features, labeled_train["sentiment"] )
```

```python
num_test_reviews = len(test["review"])
clean_test_reviews = []

print("Cleaning and parsing the test set movie reviews...\n")
for i in range(0,num_test_reviews):
    if( (i+1) % 1000 == 0 ):
        print("Review %d of %d\n" % (i+1, num_test_reviews))
    clean_test_reviews.append( review_to_words( test["review"][i] ) )

# Get a bag of words for the test set, and convert to a numpy array
test_data_features = vectorizer.transform(clean_test_reviews)
test_data_features = test_data_features.toarray()

# Use the random forest to make sentiment label predictions
result = forest.predict(test_data_features)

# Copy the results to a pandas dataframe with an "id" column and
# a "sentiment" column
output = pd.DataFrame( data={"id":test["id"], "sentiment":result} )

# Use pandas to write the comma-separated output file
output.to_csv( "word2vec/Bag_of_Words_model.csv", index=False, quoting=3 )
```
